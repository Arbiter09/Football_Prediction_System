CREATING BASIC PREDICTORS AS A BASE

***We'll create a Venue Code which basically converts the string "Home" or "Awsy" into a numerical code so that our algorithm can work on them and include the factor of "Home Ground Advantage"***

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes         #Away:0 and Home:1
```

```python
matches.head()
```

***Similarily, we will also convert the opposition team names into numerical values as well so that we can factor them into our algorithm***

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches.head()
```

***Some teams may play better at a particular time and we need to include that as well. We will be writing a Regex to extract the hour from the time column as the minutes part does not make a huge difference to our model. For eg: If the kick-off time is "16:30" we well extract the "16" part and discard the ":30" part.***

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches.head()
```

***We will encode the days into numbers as well. If you follow the EPL closely, you'll know that during the winters, there is an overload of matches and hence each team has to play twice a week, once on Wednesday/Thursday and once on Saturday/Sunday. This may affect the performance of certain teams and that is why it becomes an important feature for us to consider.***

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches.head()
```

```python
matches["target"] = (matches["result"] == "W").astype("int")    #Establishing our target variable where '0' is a loss and '1' is a win
```

```python
matches.head()
```

CREATING BASIC MACHINE LEARNING MODEL

```python
from sklearn.ensemble import RandomForestClassifier
```

***We will be using a Random Forest Classifier as it is a Machine Learning Model famously known for its ability to pickup the non-lineraity in the data. As you can observe in the data, the "opp_code" has numbers from 0-19, but that does not imply that the opposition with the code 18 is more difficult than the opposition with the code 5. Now this can be picked up by the Non-Linear ML models but cannot be identified by the Linear ML models.***

```python
rf = RandomForestClassifier(n_estimators=200, min_samples_split=10, random_state=1)
```

***The "n_estimators" parameter basically defines the number of individual decision trees but each tree has slightly different parameters. Basically, higher the number more the time it will take to train the model but will also ensure a higher accuracy. The "min_samples_split" parameter defines the minimum number of ssamples we want to have in a leaf node of a decision tree before we actually split it. The bhigher this is we will be less likely to overfit but also have a lower accuracy on the training dataset, so we'll need to experiment with it's value.***

```python
train = matches[matches["date"] < "2022-01-01"]    #subsetting our training data from the original data
```

```python
test = matches[matches["date"] > "2022-01-01"]   #subsetting our test data from the original data
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]  #list of all our defined predictors
```

```python
rf.fit(train[predictors], train["target"])
```

***We have fit our model with the training dataset. In the rf.fit(x, y), the x part suggest on which data we want the model to train on and the y part suggests what is the target that we need to predicr with the help our training data.***

```python
preds = rf.predict(test[predictors])
```

***To determine the accuracy of the matrix, we will be importing a couple of different matrics and then figure out which one works best for our case.***

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

IMPROVING PRECISION WITH ROLLING AVERAGES

```python
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City")
```

```python
group
```

***So, basically what we did here is that we grouped all the matches played by a single team into one group so that when we make prediction we can look at their past matches, shots on target, goals scored goals conceded and many more factors. This helps us include the sports term "run-of-form" into our algorithm. This is a very important feature as a team might not be that good on paper, but it may have a good run of form which heavily influences the performance of that team.***

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)  #This is for Man City only as we had called the group function for ManCity above
```

***As we can observe from the Manchester City group, we have added the rolling averages and have a lot of additional information about the team's  performance and will be able to make better predictions. So, we will now group all the other teams and appy the same rolling_averages function to their groups using lambda functions.***

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("team")
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

***So what we basically did here is that we assigned a unique id to all the rows so tha we don not have to deal with redundant row indexes which makes it easier for us to call and handle the rows.***

```python
matches_rolling
```

RETRAINING OUR MODEL

```python
def make_predictions(data, predictors):
    train = data[data["date"] < "2022-01-01"]
    test = data[data["date"] > "2022-01-01"]
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
precision
```
